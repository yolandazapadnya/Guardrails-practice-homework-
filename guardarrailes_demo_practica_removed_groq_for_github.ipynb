###   Clase Clasificador + Groq

```python
%pip install --upgrade pip
%pip install tensorflow
%pip install groq
%pip install transformers
%pip install tensorflow
%pip install torch
%pip install accelerate
%pip install torch torchvision torchaudio" 
%pip install tf-keras
%pip install guardrails-ai
%guardrails configure

from groq import Groq
from transformers import pipeline

# Verify TensorFlow installation
try:
    import tensorflow as tf
    print(f"TensorFlow version: {tf.__version__}")
except ImportError:
    print("TensorFlow is not installed correctly.")

# Verify PyTorch installation
try:
    import torch
    print(f"PyTorch version: {torch.__version__}")
except ImportError:
    print("PyTorch is not installed correctly.")

class GroqGuardrails:
    def __init__(self, groq_api_key, categories):
        self.groq_api_key = groq_api_key
        self.categories = categories
        self.classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
        self.groq_client = Groq(api_key=self.groq_api_key)

    def ask_groq(self, prompt, groq_model = "llama-3.3-70b-versatile", threshold = 0.8, categories = None):
        categories = self.categories if categories is None else categories
        classification_result = self.classifier(prompt, categories)
        print(f'\nResultados clasificación: {classification_result}')
        if any(score >= threshold for score in classification_result['scores']):
            chat_completion = self.groq_client.chat.completions.create(
                messages=[{"role": "user", "content": prompt}],
                model=groq_model,
                temperature=0,
            )
            return chat_completion.choices[0].message.content
        else:
            return "Lo siento pero no puedo responderte a eso."
```
